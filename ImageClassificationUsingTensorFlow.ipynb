## Image Classification using Tensorflow

```python
import os

empty_dataset_path = "/content/drive/MyDrive/data/parking_spot_data/cropped_dataset/empty"
non_empty_dataset_path = "/content/drive/MyDrive/data/parking_spot_data/cropped_dataset/non-empty"

#listing the dataset contents
empty_dataset_contents = os.listdir(empty_dataset_path)
non_empty_dataset_contents = os.listdir(non_empty_dataset_path)

empty_dataset_contents
```

```python
#using imageDatagenerator
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#defining the data augmentation params
data_gen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range = 0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
import cv2
```

```python
empty_images_li = []
for i in empty_dataset_contents:
  print(i)
  image = cv2.imread(f"/content/drive/MyDrive/data/parking_spot_data/cropped_dataset/empty/{i}")
  empty_images_li.append(image)
empty_images_li
```

```python
non_empt_images_li = []
for i in non_empty_dataset_contents:
  print(i)
  image = cv2.imread(f"{non_empty_dataset_path}/{i}")
  non_empt_images_li.append(image)
non_empt_images_li
```

```python
print(len(empty_images_li))
print(len(non_empt_images_li))
```

```python
import numpy as np
```

```python
#applying the data augmentation on empty dataset
augmented_count = 0
for i in empty_images_li:
  for j in range(10):
    augmented_image = next(data_gen.flow(np.expand_dims(i, axis=0), batch_size=1))[0]
    augmented_count += 1
    cv2.imwrite(f"/content/drive/MyDrive/data/parking_spot_data/augmented_dataset/empty/empty_{augmented_count}.jpg", augmented_image)
print("Total images:",augmented_count)
```

```python
#applying non empty dataset
augmented_count = 0
for i in non_empt_images_li:
  for j in range(10):
    augmented_image = next(data_gen.flow(np.expand_dims(i, axis=0), batch_size=1))[0]
    augmented_count += 1
    cv2.imwrite(f"/content/drive/MyDrive/data/parking_spot_data/augmented_dataset/non-empty/non_empty_{augmented_count}.jpg", augmented_image)
print("Total images:",augmented_count)
```

```python
#loading the dataset
dataset_path = "/content/drive/MyDrive/data/parking_spot_data/augmented_dataset"
classes = ["empty", "non-empty"]
```

```python
img_height = 32
img_width = 32
```

```python
batch_size = 32
split_ratio = 0.2
```

```python
#creating imagedatagenerator
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=split_ratio
)
```

```python
#training data genrator
train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode="categorical",
    subset="training"
)
```

```python
#validation data generator
validation_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode="categorical",
    subset="validation"
)
```

```python
#building model
#importing packages
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
```

## CNN Model Structure

<img src="https://learnopencv.com/wp-content/uploads/2023/01/tensorflow-keras-cnn-architecture.png">

```python
def cnn_model(input_shape=(32, 32, 3)):
  model = Sequential()
  #Conv Block 1 for 32 filters
  model.add(Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=input_shape))
  model.add(Conv2D(filters=32, kernel_size=3, padding="same", activation="relu"))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  #conv Block 2 for 64 filters
  model.add(Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
  model.add(Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  #conv block 3: 64 filters
  model.add(Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
  model.add(Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  #flatten the convolutional features
  model.add(Flatten())
  model.add(Dense(512, activation="relu"))
  model.add(Dense(3, activation="softmax"))
  return model
```

```python
#creating the model
model = cnn_model()
model.summary()
```

```python
#compiling the model
model.compile(
    optimizer="rmsprop",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
#training the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples//batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples//batch_size,
    epochs=10
)
```

```python
#evaluate the model on the validation
loss, accuracy = model.evaluate(validation_generator)
print("validation accuracy:", accuracy)
```

```python
import numpy as np
```

```python
#getting a batch of validation data
x_batch, y_batch = next(validation_generator)
#make predictions on the batch
predictions = model.predict(x_batch)
#getting the predicted class for each image
predicted_classes = np.argmax(predictions, axis=1)
#actual_class of the each image
actual_classes = np.argmax(y_batch, axis=1)
#printing the predicted and actual class
for i in range(10):
  print("predicted_class:", predicted_classes[i])
  print("actual_class:", actual_classes[i])
```
