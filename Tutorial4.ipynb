```python
#import required packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_image = cv2.imread("/content/drive/MyDrive/AbelFolder/last_supper.jpg")
orginal_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.imshow(orginal_image_rgb)
```

```python
#cv2.threshold - args = src image, thresh value, max_value, threshtype
#cv2.adaptivethreshold
#OCR =
#global threshold
#gray image
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
retval, last_supper_global_thresh = cv2.threshold(gray_image, 50, 255, cv2.THRESH_BINARY)
plt.imshow(last_supper_global_thresh, cmap="gray")
```

```python
plt.imshow(gray_image, cmap="gray")
```

```python
#global threshold, 130, 255
ret, global_threshold_2 = cv2.threshold(gray_image, 130, 255, cv2.THRESH_BINARY)
plt.imshow(global_threshold_2, cmap="gray")
```

```python
#adaptive threshold
last_supper_adaptive = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 7)
plt.imshow(last_supper_adaptive, cmap="gray")
```

```python
#bitwise operations
#and, or, xor, not
#orginal_images
anime_img = cv2.imread("/content/drive/MyDrive/AbelFolder/anime_image.jpg", cv2.IMREAD_COLOR)
warrior_img = cv2.imread("/content/drive/MyDrive/AbelFolder/warrior.png", cv2.IMREAD_COLOR)
anime_img_rgb = cv2.cvtColor(anime_img, cv2.COLOR_BGR2RGB)
warrior_img_rgb = cv2.cvtColor(warrior_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(anime_img_rgb);plt.title("Anime")
plt.subplot(122);plt.imshow(warrior_img_rgb);plt.title("Warrior")
```

```python
print(anime_img_rgb.shape)
print(warrior_img_rgb.shape)
```

```python
anime_img_resized = cv2.resize(anime_img_rgb, (warrior_img_rgb.shape[1], warrior_img_rgb.shape[0]))
```

```python
print(anime_img_resized.shape)
```

```python
#bitwise AND
result_and = cv2.bitwise_and(anime_img_resized, warrior_img_rgb, mask=None)
plt.imshow(result_and)
```

```python
#bitwise or
result_or = cv2.bitwise_or(anime_img_resized, warrior_img_rgb, mask=None)
plt.imshow(result_or)
```

```python
#XOR
result_xor = cv2.bitwise_xor(anime_img_resized, warrior_img_rgb, mask=None)
plt.imshow(result_xor)
```

```python
#manipulation application
#image - with a ABEL
#image - photo
abel_image = cv2.imread("/content/drive/MyDrive/AbelFolder/ABELANAND.png")
abel_img_rgb = cv2.cvtColor(abel_image, cv2.COLOR_BGR2RGB)
plt.imshow(abel_img_rgb)
```

```python
print(abel_image.shape)
```

```python
bg_img = cv2.imread("/content/drive/MyDrive/AbelFolder/anime_image.jpg")
bg_img_rgb = cv2.cvtColor(bg_img, cv2.COLOR_BGR2RGB)
plt.imshow(bg_img_rgb)
```

```python
print(bg_img.shape)
```

```python
bg_img_resize = cv2.resize(bg_img_rgb, (abel_img_rgb.shape[1], abel_img_rgb.shape[0]))
```

```python
print(bg_img_resize.shape)
```

```python
#create a mask for our original image
abel_img_gray = cv2.cvtColor(abel_img_rgb, cv2.COLOR_RGB2GRAY)
ret, mask_img = cv2.threshold(abel_img_gray, 177, 255, cv2.THRESH_BINARY)
plt.imshow(mask_img, cmap="gray")
```

```python
#inverting the image using not operator
mask_inv = cv2.bitwise_not(mask_img)
plt.imshow(mask_inv, cmap="gray")
```

```python
#overlapping using mask
result_img = cv2.bitwise_and(bg_img_resize, bg_img_resize, mask=mask_img)
plt.imshow(result_img)
```

```python
#overlapping method
result_img2 = cv2.bitwise_and(abel_img_rgb, abel_img_rgb, mask=mask_inv)
plt.imshow(result_img2)
```

```python
#for merging add(cv2)
final = cv2.add(result_img, result_img2)
plt.imshow(final)
```

```python
cv2.imwrite("abel_img.png", final[:, :, ::-1])
```
